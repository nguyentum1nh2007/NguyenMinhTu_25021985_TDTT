```python
def pretty_array(lst: list) -> str:
    check = "YES"
    res = 0
    cont = set(lst)
    for i in cont:
        if lst.count(i) > i:
            res = res + lst.count(i) - i
            check = "NO"
        elif lst.count(i) < i:
            res += lst.count(i)
            check = "NO"
        else: continue
    if check == "YES": return check
    else: return check + " " + str(res)
a_16 = list(map(int, input().split()))
pretty_array(a_16)
```

```python
def freq_above_half(lst: list) -> int:
    cont = set(lst)
    para = len(lst)/2
    for i in cont:
        if lst.count(i) > para:
            return i
a_17 = list(map(int, input().split()))
freq_above_half(a_17)
```

```python
def odd_even_seperator(lst: list) -> str:
    even = []
    odd = []
    for i in lst:
        if i % 2 == 0: even.append(i)
        elif i % 2 != 0: odd.append(i)
        else: raise Exception("Invalid Input")
    even.sort()
    odd.sort(reverse=True)
    return " ".join(map(str, even)) + " " + " ".join(map(str, odd))

a_18 = list(map(int, input().split()))
odd_even_seperator(a_18)
```

```python
def arrays_merger(arr1: list, arr2: list) -> str:
    is_descending = False
    if (len(arr1) > 1 and arr1[0] > arr1[-1]) or (len(arr2) > 1 and arr2[0] > arr2[-1]):
        is_descending = True
    res = arr1 + arr2
    if is_descending: return " ".join(map(str, sorted(res, reverse=True)))
    elif not is_descending: return " ".join(map(str, sorted(res)))
    else: raise Exception("Fatal Error")
a_19_1 = list(map(int, input().split()))
a_19_2 = list(map(int, input().split()))
arrays_merger(a_19_1, a_19_2)
```

```python
def sum_of_k_digits_numbers(lst: list, k: int) -> int:
    sum = 0
    for i in list(map(str, lst)):
        if len(i) == k: sum += int(i)
    return sum
a_20_1 = list(map(int, input().split()))
a_20_int = int(input())
sum_of_k_digits_numbers(a_20_1, a_20_int)
```

```python
def accumulated_calendar_days(n: int, t:int):
    if n > 31 or t > 12: return "invalid"
    res = 0
    month = [31,29,31,30,31,30,31,31,30,31,30,31]
    for i in range(0, t-1):
        res += month[i]
    return res+n
a_21_n, a_21_t = map(int, input().split()) 
accumulated_calendar_days(a_21_n, a_21_t)
```

```python
def most_freq_names(N: int, lst: list) -> str:
    #value list format [[0][1][2]]
    dict = {}
    for i in lst:
        if i[-1] not in dict: dict[i] = 1
        elif i[-1] in dict: dict[i] += 1
        else: raise Exception("Fatal Error")
    cont = sorted(dict.items(), key=lambda item: item[1])
    return cont[0][0]
a_22_list = []
a_22_n = int(input())
for i in range(a_22_n):
    name = str(input())
    a_22_list.append(name)
most_freq_names(a_22_n, a_22_list)
```

```python
import re
def count_names(text):
    words = re.findall(r'\b[a-zA-Z]+\b', text)
    count = 0
    for word in words:
        if len(word) >= 2 and word[0].isupper() and word[1:].islower():
            count += 1
    return count
a_23 = str(input())
```

```python
def remove_consecutive_duplicates(text):
    words = text.split()
    if not words:
        return ""
    result = [words[0]]
    for i in range(1, len(words)):
        if words[i] != words[i-1]:
            result.append(words[i])
    return " ".join(result)
a_24 = str(input())
```

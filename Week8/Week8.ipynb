```python
def binary_search(array, left, right, target):
    while left <= right:
        mid = (left + right) // 2

        if array[mid] == target:
            return mid

        elif array[mid] < target:
            left = mid + 1

        else:
            right = mid - 1
    return -1

lst = list(map(int, input().split()))
x = int(input())
binary_search(lst, 0, len(lst)-1, x)
```

```python
def bubble_sort(lst):
    n = len(lst)
    for i in range(n-1):
        swapped = False
        for j in range(n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
                swapped = True
        if not swapped:
            break
def selectionSort(array, size):
    for ind in range(size):
        min_index = ind

        for j in range(ind + 1, size):
            if array[j] < array[min_index]:
                min_index = j

        array[ind], array[min_index] = array[min_index], array[ind]
```

```python
def most_frequent(List):
    counter = 0
    num = List[0]
    
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i

    return num, counter
```

```python
def count_pairs_with_sum(nums, target_x):
    count = 0
    seen = {}  

    for num in nums:
        complement = target_x - num
        if complement in seen:
            count += seen[complement]
        seen[num] = seen.get(num, 0) + 1
        
    return count
```

```python
def list_flatter(nested_list):
    result = []
    
    for i, sublist in enumerate(nested_list):
        if i % 2 == 1:
            result.extend(sublist[::-1])
        else:
            result.extend(sublist)
            
    return result
```

```python
def length_of_lis(nums):
    if not nums:
        return 0
    dp = [1] * len(nums)
    
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
                
    return max(dp)
```

```python
import heapq

def min_interval_length(intervals, queries):
    sorted_queries = sorted((q, i) for i, q in enumerate(queries))
    
    intervals.sort(key=lambda x: x[0])
    
    min_heap = []  
    result = [-1] * len(queries)
    i = 0          
    for q, idx in sorted_queries:
        while i < len(intervals) and intervals[i][0] <= q:
            l, r = intervals[i]
            length = r - l + 1
            heapq.heappush(min_heap, (length, r))
            i += 1
        
        while min_heap and min_heap[0][1] < q:
            heapq.heappop(min_heap)
        
        if min_heap:
            result[idx] = min_heap[0][0]
        else:
            result[idx] = -1
            
    return result
```

```python
def count_smaller(nums):
    n = len(nums)
    if not n:
        return []
    
    indices = list(range(n))
    counts = [0] * n
    
    def merge_sort(arr):
        if len(arr) <= 1:
            return arr
        
        mid = len(arr) // 2
        left = merge_sort(arr[:mid])
        right = merge_sort(arr[mid:])
        
        sorted_arr = []
        i = j = 0
        num_right_smaller = 0
        
        # Merge step
        while i < len(left) and j < len(right):
            if nums[left[i]] > nums[right[j]]:
                num_right_smaller += 1
                sorted_arr.append(right[j])
                j += 1
            else:
                counts[left[i]] += num_right_smaller
                sorted_arr.append(left[i])
                i += 1
        
        while i < len(left):
            counts[left[i]] += num_right_smaller
            sorted_arr.append(left[i])
            i += 1
        
        if j < len(right):
            sorted_arr.extend(right[j:])
            
        return sorted_arr

    merge_sort(indices)
    return counts
```

```python
import bisect

def max_subarray_sum_modulo(nums, m):
    sorted_prefixes = [0]
    current_sum = 0
    max_mod_sum = 0
    
    for num in nums:
        current_sum = (current_sum + num) % m

        max_mod_sum = max(max_mod_sum, (current_sum - sorted_prefixes[0] + m) % m)
        
        idx = bisect.bisect_right(sorted_prefixes, current_sum)
        
        if idx < len(sorted_prefixes):
            candidate = (current_sum - sorted_prefixes[idx] + m) % m
            max_mod_sum = max(max_mod_sum, candidate)
            
        bisect.insort(sorted_prefixes, current_sum)
        
    return max_mod_sum
```

```python
def solve_fastest_finger():
    labels = ['A', 'B', 'C', 'D']
    options = []

    print("Enter the 4 options (press Enter after each):")

    for label in labels:
        text = input(f"{label}: ").strip()
        options.append((label, text))

    sorted_options = sorted(options, key=lambda x: x[1].lower())

    result_order = [item[0] for item in sorted_options]

    print("\nResult:")
    print(" ".join(result_order))
```

```python
from collections import Counter

def find_lhs(nums):
    counts = Counter(nums)
    max_length = 0

    for num in counts:
        if num + 1 in counts:
            current_length = counts[num] + counts[num + 1]
            max_length = max(max_length, current_length)

    return max_length
```

```python
def find_number_seven(nums):
    indices = [index for index, value in enumerate(nums) if value == 7]
    if not indices:
        return "Not found"
    
    indices.sort(reverse=True)
    return " ".join(map(str, indices))
```

```python
def solve_nemo_neighbors(guests):
    n = len(guests)
    
    try:
        idx = guests.index("Nemo")
    except ValueError:
        return "Nemo is not in the list"
    prev_guest = guests[(idx - 1) % n]
    next_guest = guests[(idx + 1) % n]
    
    return f"{prev_guest} and {next_guest}"
```

```python
def merge_lists(list1, list2):
    merged = []
    i = 0
    j = 0
    
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j] + 2:
            merged.append(list1[i])
            i += 1
        else:
            while j < len(list2) and list2[j] < list1[i]:
                merged.append(list2[j])
                j += 1
            if i < len(list1):
                merged.append(list1[i])
                i += 1
                
    if i < len(list1):
        merged.extend(list1[i:])
        merged.sort()
    if j < len(list2):
        merged.extend(list2[j:])
    return merged
        


list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))

print(merge_lists(list1, list2))
```

```python
def count_less_equal(matrix, target):
    n = len(matrix)
    m = len(matrix[0])
    count = 0
    row = n - 1
    col = 0
    
    while row >= 0 and col < m:
        if matrix[row][col] <= target:
            count += (row + 1)
            col += 1
        else:
            row -= 1
            
    return count

def find_kth_smallest(matrix, k):
    if not matrix or not matrix[0]:
        return None

    n = len(matrix)
    m = len(matrix[0])
    
    low = matrix[0][0]
    high = matrix[n - 1][m - 1]
    
    while low < high:
        mid = (low + high) // 2

        count = count_less_equal(matrix, mid)
        
        if count < k:
            low = mid + 1
        else:
            high = mid
            
    return low
        

data_matrix = []
while True:
    try:
        line = input()
        if not line:
            break
        row = [int(x) for x in line.split()]
        if len(row) == 1: 
            n = int(row[0])
            break
        data_matrix.append(row)
    except EOFError:
        break
    except ValueError:
        print("Invalid input: please enter only space-separated numbers.")
        continue
# n = int(input())
print(find_kth_smallest(data_matrix, n))
```

```python
def strings_solver():
    a = input().strip()
    b = input().strip()
    n, m = len(a), len(b)
    
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True

    for i in range(n):
        for j in range(m + 1):
            if not dp[i][j]:
                continue
            
            curr_a = a[i]
            if j < m and curr_a.upper() == b[j]:
                dp[i+1][j+1] = True
            
            if curr_a.islower():
                dp[i+1][j] = True

    print("YES" if dp[n][m] else "NO")

n = int(input())
try:
    for _ in range(n):
        strings_solver()
except EOFError:
    pass
```

```python
def check_parent(lst1: str, lst2: str) -> str:
    lst1 = "".join(lst1.split())
    lst2 = "".join(lst2.split())
    if lst2.find(lst1) == -1:
        return "NO"
    else: return "YES"
lst1 = str(input())
lst2 = str(input())
print(check_parent(lst1, lst2))
```

```python
import sys

def next_permutation(n, nums):
    i = n - 2
    while i >= 0 and nums[i] >= nums[i+1]:
        i -= 1
    
    if i >= 0:
        j = n - 1
        while nums[j] <= nums[i]:
            j -= 1
        nums[i], nums[j] = nums[j], nums[i]
    left = i + 1
    right = n - 1
    while left < right:
        nums[left], nums[right] = nums[right], nums[left]
        left += 1
        right -= 1
        
    return nums

def solve():
    input_data = sys.stdin.read().split()
    if not input_data:
        return
    iterator = iter(input_data)
    try:
        n = int(next(iterator))
        nums = []
        for _ in range(n):
            nums.append(int(next(iterator)))
            
        result = next_permutation(n, nums)
        print(*(result))
        
    except StopIteration:
        pass
```

```python
def canPlaceFlowers(flowerbed, k):
    spot = 0
    for i in range(0, len(flowerbed), 2):
        if flowerbed[i] == 0:
            spot += 1
    if spot >= k: return True
    else: return False

flowerbed = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    
    for move in moves:
        if move == 'U':   
            y += 1
        elif move == 'D':
            y -= 1
        elif move == 'R': 
            x += 1
        elif move == 'L': 
            x -= 1

        if (x, y) in visited:
            return True

        visited.add((x, y))

    return False
moves = str(input())
print(is_path_crossing(moves))
```

```python
from bisect import bisect_left

def findRadius(houses, heaters):
    heaters.sort()
    
    min_radius = 0
    
    for house in houses:
        i = bisect_left(heaters, house)
        left_dist = float('inf')

        if i > 0:
            left_dist = house - heaters[i - 1]
        right_dist = float('inf')

        if i < len(heaters):
            right_dist = heaters[i] - house
        closest_dist = min(left_dist, right_dist)
        
        min_radius = max(min_radius, closest_dist)
        
    return min_radius
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```

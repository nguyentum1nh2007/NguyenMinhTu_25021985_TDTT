```python
def Max_Of_Two(a,b):
    return max(a,b)
Max_Of_Two(2,4)
```

```python
def swap(a,b):
    return tuple[b,a]
swap(2,4)
```

```python
def is_prime_number(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
            break
    return True
is_prime_number(3)
```

```python
def is_perfect_number(n):
    divisors = list(filter(lambda x: n % x == 0, range(1, n)))
    if sum(divisors) == n:
        return True
    else:
        return False
is_perfect_number(28)
```

```python
def first_instance(list, n):
    for index, item in enumerate(list):
        if item == n:
            return index
    return -1
list = [1,2,3,4,5,1,7]
first_instance(list, 6)
```

```python
def factorial_calculator(n):
    res = 1
    for i in range(2, n+1):
        res *= i
    return res
factorial_calculator(3)
```

```python
def calculator(a, b, op):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a*b
    elif op == "/":
        return a/b
    else: return "Syntax Error"
calculator(2,3,"+")
```

```python
def hamming_distance_integers(a: int, b: int) -> int:
    xor_result = a ^ b
    return bin(xor_result).count('1')
hamming_distance_integers(4,1)
```

```python
def sum_of_intergers(n: int) -> int:
    sum = 0
    for i in str(n):
        sum += int(i)
    return sum
sum_of_intergers(2234)
```

```python
def check_isomorphic(w1, w2):
    map_w1_to_w2 = {}
    map_w2_to_w1 = {}
    for char1, char2 in zip(w1, w2):
        
        if char1 in map_w1_to_w2:
            if map_w1_to_w2[char1] != char2:
                return False  
        
        else:
            if char2 in map_w2_to_w1:
                return False  
            
            map_w1_to_w2[char1] = char2
            map_w2_to_w1[char2] = char1

    return True
check_isomorphic("axbxx", "cafaa")
```

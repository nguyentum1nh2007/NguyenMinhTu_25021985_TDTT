```python

class Counter:
    def __init__(self):
        self.value = 0 #value laÌ€ tÃªn thuÃ´Ì£c tiÌnh cuÌ‰a lÆ¡Ìp Counter

    def increase(self): #increase laÌ€ tÃªn phÆ°Æ¡ng thÆ°Ìc cuÌ‰a lÆ¡Ìp Counter
        self.value += 1

c = Counter() #DuÌ€ng haÌ€m taÌ£i __init__ de taÌ£o ra Ä‘Ã´Ìi tÆ°Æ¡Ì£ng c
c.increase() # GoÌ£i tÆ¡Ìi phÆ°Æ¡ng thÆ°Ìc increase tÆ°Ì€ Ä‘Ã´Ìi tÆ°Æ¡Ì£ng c
c.increase()
print(c.value)
```


## 5. CÃ¡c phÆ°Æ¡ng thá»©c Ä‘áº·c biá»‡t (__dunder__ methods)
#### ÄoÌ£c thÃªm bÃªn file kia :D 


```python

class Point:
    def __init__(self, _x, _y): # haÌ€m - phÆ°Æ¡ng thÆ°Ìc khÆ¡Ì‰i taÌ£o
        self.x = _x
        self.y = _y

    def __str__(self): #haÌ€m - phÆ°Æ¡ng thÆ°Ìc in (print) Ä‘Ã´Ìi tÆ°Æ¡Ì£ng
        return f"({self.x}, {self.y})"

    def __add__(self, other):# HaÌ€m - phÆ°Æ¡ng thÆ°Ìc cÃ´Ì£ng Ä‘Ã´Ìi tÆ°Æ¡Ì£ng vÆ¡Ìi 1 Ä‘Ã´Ìi tÆ°Æ¡Ì£ng khaÌc
        return Point(self.x + other.x, self.y + other.y)
    
    def __del__ (self):#HaÌ€m - PhÆ°Æ¡ng thÆ°Ìc huÌ‰y Ä‘Ã´Ìi tÆ°Æ¡Ì£ng - TÆ°Ì£ Ä‘Ã´Ì£ng Ä‘Æ°Æ¡Ì£c goÌ£i khi kÃªÌt thuÌc
        print("Object deleted")

p1 = Point(1, 2)
p2 = Point(3, 4)
print(p1)
print(p1 + p2)
```


## 6. Káº¿ thá»«a (Inheritance)


```python

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

d = Dog()
d.speak()
```


## 7. super()


```python

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name) #goÌ£i haÌ€m taÌ£o cuÌ‰a lÆ¡Ìp Animal
        self.breed = breed

dog = Dog("Lucky", "Husky")
print(dog.name, dog.breed)
```


## 8. Encapsulation â€“ ÄÃ³ng gÃ³i (_ vÃ  __)


```python

class Account:
    def __init__(self, balance):
        self._balance = balance

    def get_balance(self):
        return self._balance

acc = Account(1000)
print(acc.get_balance())
```


## 9. Property (@property)


```python

class Circle:
    def __init__(self, r):
        self._r = r

    @property
    def area(self):
        return 3.14 * self._r ** 2

c = Circle(5)
print(c.area)
```


## 10. Classmethod & Staticmethod


```python

class MathUtils:
    pi = 3.14

    @classmethod
    def circle_area(cls, r):
        return cls.pi * r * r

    @staticmethod
    def add(a, b):
        return a + b

print(MathUtils.circle_area(3))
print(MathUtils.add(2, 5))
```


## 12. VÃ­ dá»¥ tá»•ng há»£p (Mini Project)


```python

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("KhÃ´ng Ä‘á»§ tiá»n")

    def __str__(self):
        return f"{self.owner}: {self.balance}"

acc = BankAccount("Long", 1000)
acc.deposit(500)
acc.withdraw(300)
print(acc)
```

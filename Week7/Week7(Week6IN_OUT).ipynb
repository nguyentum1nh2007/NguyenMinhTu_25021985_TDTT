```python
def remove_dupes(lst: list) -> list:
    res = set(lst)
    return list(res)
l = [1,3,1,5,6]
remove_dupes(l)
```

```python
def accumulated_sum(lst: list) -> list:
    sum = 0
    res = []
    for i in lst:
        sum += i
        res.append(sum)
    return res
a = [1,1,2,4,5]
accumulated_sum(a)
```

```python
def spin_tuple(tup, k):
    n = len(tup)
    x = k % n
    first = tup[:x]
    last = tup[x:]
    res = last + first
    
    return res
tup = (4,5,6,7)
spin_tuple(tup, 2)
```

```python
def create_index_map():
    text = input("Nhập chuỗi (các từ cách nhau bởi dấu cách): ")
    
    words = text.split()
    index_map = {}
    for index, word in enumerate(words):
        if word in index_map:
            index_map[word].append(index)
        else:
            index_map[word] = [index]
    for word in index_map:
        index_map[word] = tuple(index_map[word])
    print(index_map)
create_index_map()
```

```python
def numb_dict(list):
    positives = []
    negatives = []
    zeros = []
    for i in list:
        if i > 0: positives.append(i)
        elif i < 0: negatives.append(i)
        else: zeros.append(i)
    dict = {}
    dict["positives"] = positives
    dict["negatives"] = negatives
    dict["zeros"] = zeros
    return dict
numb_dict(l)
```

```python
def count(list):
    dict = {}
    for i in list:
        if i not in dict:
            dict[i] = 1
        else:
            dict[i] = dict[i] + 1
    return dict
count(l)
```

```python
def concat_dictionary(dict1, dict2):
    res = dict1.copy()

    for key, value in dict2.items():

        if key in res:
            res[key] += value
        else:
            res[key] = value

    keys = sorted(res.keys())
    dict = {key: res[key] for key in keys}
    
    return dict

def string_to_dict(s):
    d = {}
    try:
        if not s.strip():
            return {}
        pairs = s.split(',')
        for pair in pairs:
            key, value_str = pair.split(':')
            d[key.strip()] = int(value_str.strip())
    except Exception as e:
        raise ValueError(f"{e}")
    return d
```

```python
def find_sum_pairs(k, data_list):
    n = len(data_list)
    res = []
    
    for i in range(n):
        for j in range(i + 1, n):
            num1 = data_list[i]
            num2 = data_list[j]

            if num1 + num2 == k:
                if num1 != num2:
                    pair = (num1, num2)
                    pair = sorted(pair)
                    res.append(pair)
    res.sort()
    return res
lst = [1,56,74,6,3,2]
find_sum_pairs(5, lst)
```

```python
def reverse_dict(dict):
    res = {}
    for i in dict:
        res[dict[i]] = i
    return res 
```

```python
def find_equal_numbs(list1, list2):
    res = []
    for i in list1:
        if i in list2:
            res.append(i)
    return res
```

```python
def filtered_dict(dict, k):
    res = {}
    for i in dict:
        if dict[i] > k:
            res[i] = dict[i]
    return res
```

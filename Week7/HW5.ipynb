```python
def matrix_modifier(m, n):
    matrix = []
    child = []
    for i in range(n):
        for index in range(m):
            x = int(input("Nhap: "))
            child.append(x)
        matrix.append(child)
        child = []
    return matrix
def matrix_calc(k, m, n):
    mat =  matrix_modifier(m,n)
    print("Tong hang:", sum(mat[k-1]))
    cont = 0
    for i in range(n):
        cont += mat[i][k-1]
    print("Tong cot:", cont)
matrix_calc(1, 2, 2)
```

```python
def greatest_number_index(list):
    max = 0
    ind = 0
    for index, i in enumerate(list):
        if i >= max:
            max = i
            ind = index
    print(ind)
lst = [1,3,6,7,5,7]
greatest_number_index(lst)
```

```python
def create_index_map():
    text = input("Nhập chuỗi (các từ cách nhau bởi dấu cách): ")
    
    words = text.split()
    index_map = {}
    for index, word in enumerate(words):
        if word in index_map:
            index_map[word].append(index)
        else:
            index_map[word] = [index]
    for word in index_map:
        index_map[word] = tuple(index_map[word])
    print(index_map)
create_index_map()
```

```python
def zero_mover(lst):
    count = 0
    for index, i in enumerate(lst):
        if i == 0:
            count += 1
            lst.pop(index)
    for i in range(count):
        lst[:0] = [0]
    return lst
l = [1,0,4,0,6,5]
zero_mover(l)
```

```python
def missing_numbers(lst):
    cont = []
    for i in range(0,10):
        if i not in lst:
            cont.append(i)
    return cont
missing_numbers(l)
```

```python
def sum_of_local_maxima(lst):
    sum = 0
    for i in range(0, len(lst)-1):
        if lst[i] > lst[i+1] and lst[i] > lst[i-1]:
            sum += lst[i]
    return sum
test = [1,4,5,3,6]
sum_of_local_maxima(test)
```

```python
def filter(lst, x):
    res = []
    for i in lst:
        if i < x:
            res.append(i)
    if len(res) == 0: return -1
    else: return res
filter(l, 5)
```

```python
def list_tendency(lst):
    increase =  lst.sort()
    decrease = lst.sort(reverse=True)
    if lst == increase: return "increase"
    elif lst == decrease: return "decrease"
    else: return "unknown"
list_tendency(l)
```
